```python
def softmax(x):
  return np.exp(x*.5)/sum(np.exp(x*.5))
```

```python
state = 'ee89d'
history[history['state']== state]
```

```python
state = 'ee89d'
rewards = history[history['state']== state].groupby('action').mean()['reward']
# softmax(rewards)
rewards
```

```python
softmax(rewards)
```

```python
# from scipy.special import softmax

for index, row in submit.iterrows():
    state = row['state']
    rewards = history[history['state']== state].groupby('action').mean()['reward']
    try:
      softmax_  =  softmax(rewards)
      submit.loc[submit.state == state, 'action0'] = int(100 * softmax_[0])
      submit.loc[submit.state == state, 'action1'] = int(100 * softmax_[1])
      submit.loc[submit.state == state, 'action2'] = int(100 * softmax_[2])
    except:
      pass
```
